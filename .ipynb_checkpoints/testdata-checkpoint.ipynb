```python
from random import randint
import pandas as pd
```

```python
def generateMachines(number):
    machines = []
    for i in range(1,number+1):
        machines.append(f'Machine_{i}')
    return machines
```

```python
def generateDataObjects(number_machines,observations):
    machines = generateMachines(number_machines)
    observationList = []
    for i in range(0,observations):
        machine = randint(0,number_machines-1)
        observation = {"id":i,"Machine":machines[machine],"Week":randint(1,10),"Labor Hours":randint(1,5),"Downtime Hours":randint(1,5)}
        observationList.append(observation)
    return observationList
```

```python
def generateDataFrame(number_machines,observations):
    data = generateDataObjects(number_machines,observations)
    testData = pd.DataFrame(data)
    return testData
```

```python
generateDataFrame(5,20)
```

```python
TestData = generateDataFrame(5,20)
```

```python
TestData.append({"id":20,"Machine":"Machine_1","Week":1,"Labor Hours":0,"Downtime Hours":0}, ignore_index = True)
```

```python
#TestData.groupby(["Machine","Week"]).agg({"Labor Hours":['sum'],"Downtime Hours":['sum']})
machines = set(TestData["Machine"])
weeks = set(TestData["Week"])
```

```python
machineArray = []
for machine in machines:
    machineData = TestData[TestData['Machine']==machine]
    machineArray.append(machineData)
```

```python
def fillgaps(dataFrame,machine_name):
    currId = dataFrame["id"].max()
    print(currId)
    #result = dataFrame.copy()
    arrayGaps = []
    for week in weeks:
        flag = dataFrame[dataFrame["Week"]==week]["Week"].count()
        if flag == 0:
            line = {"id":currId+1,"Machine":machine_name,"Week":week,"Labor Hours":0,"Downtime Hours":0}
            print(line)
            arrayGaps.append(line)
            currId = currId +1
    result = dataFrame.append(pd.DataFrame(arrayGaps))
    return result
```
